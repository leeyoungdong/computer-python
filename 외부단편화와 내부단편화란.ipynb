# **외부 단편화와 내부 단편화란**

메모리 단편화(Memory Fragmentation)는 RAM에서 메모리의 공간이 작은 조각으로 나눠져 사용가능한 메모리가 존재하지만, 할당이 불가능한 상태. 어떤 Task가 메모리를 할당받기 위해 연속된 메모리 공간이 필요, 그러나 메모리 단편화가 일어나면 할당되지 않은 메모리의 총량이 충분해도, 이들이 작은 메모리 파편으로 존재 하고있기 때문에 메모리 할당이 불가능한 상황이 발생.

image.png

**내부 단편화(Internal Fragmentation)**

메모리를 할당할 때 프로세스가 필요한 양보다 더 큰 메모리가 할당되어 프로세스에서 사용하는 메모리 공간이 낭비되는 상황.

 예를 들어 메모장을 켰는데 OS가 4kb를 할당해주었는데, 실제로 1kb만 사용했다고 하면 필요이상으로 프로세스가 메모리를 할당받았다. 내부 단편화가 3kb만큼 생겼다고 할수있다. 

**외부 단편화(External Fragmentation)**

작은 메모리가 중간중간에 존재해, 메모리 공간은 충분하지만 실제로 할당할수 없는 상황을 말함. 메모리가 할당되고 해제되는 작업이 반복될 때 작은 메모리가 중간에 생기게된다. 이를 Hole이라고 부름. 이렇게 중간에끼인 Hole들이 많아지면, 메모리 낭비로 인해 총 가용공간이 충분하더라도 새로운 Task가 메모리를 할당받지 못하는 상황이 만들어짐. 

예를 들어 메모리 할당과 해제의 반복으로 Hole이 만들어졌다고 하면, hole의 총합은 16MB이고, 9MB의 프로세스를 할당하고자 한다.빈 메모리의 공간중 제일 큰 메모리hole의 크기가 9MB보다 작을때, 외부 단편화가 일어났다고 할 수 있다.

# **메모리 할당 방법**

연속 할당/ 불연속 할당박으로 나뉨. 연속할당방식에는 고정분할, 가변분할 방식이있다. 불연속 할당방식에는 페이징 기법과 세그멘테이션 기법, 페이지 세그멘테이션기법 등이있다.

**압축(Compaction)기법**

주기적으로 삭제 공간을 회수하여, 메모리 공간을 정리하는 방식. 그러나 비용이 많이든다는 단점이 존재

**페이징(Paging)기법**

가상메모리사용, 외부 단편화 해결, 내부 단편화 존재

image.png

프로세스를 일정한 단위로 잘라 사용하는 방식. 프로세스의 주소 공간을 동일한 크기의 페이지 단위로 나눠 물리적 메모리의 서로 다른 위치에 피이지들을 저장하는 방식. 논리(가상) 메모리의 단위는 page(페이지), 물리 메모리의 단위는 프레임(frame)

페이징 기법을 사용하면 연속적이지 않은 공간도 활용할 수 있기 때문에 외부 단편화 문제를 해결할 수 잇다. 대신 페이지 단위에 알맞게 꽉채워 쓰는게 아니므로, 내부 단편화 문제는 존재. 페이지 단위를 작게하면 내부 단편화 문제도 해결할 수 있지만, 대신 page mapping과정이 많아지므로 효율이 떨어짐.

**세그멘테이션(Segmentation)기법**

가상메모리사용, 내부 단편화 해결, 외부 단편화 존재재

image.png

하나의 프로세스를 구성하는 주소 공간은 코드, 데이터, 스택등의 의미있는 단위들로 구성되며 이렇게 기능적인 구조를 하나의 세그먼트 단위삼아 분할하여 저장됨.

페이징 기법에서 가상메모리를 같은 크기의 단위로 분할했지만 세그멩테이션 기법에서는 가상메모리를 서로 크기가 다른 논리적 단위인 세그먼트로 분할해 메모리를 할당하여 실제 메모리 주소로 변환한다. 각 세그먼트는 연속적인 공간에 저장되어 있다. 세그먼트들의 크기가 다르기 때문에 미리 분할해 둘 수 없고 메모리에 적재될 때 빈공간을 찾아 할당해야한다. 마찬가지로 mapping을 위해 세그먼트 테이블이 필요

프로세스가 필요한 만큼 메모리를 할당해주기 때문에 내부 단편화는 일어나지 않으나, 여전히 중간에 프로세스가 메모리를 해제하면 생기는 hole = 외부단편화 문제는 존재재
