# **Context Switching이란?**

현재 실행중인 프로세스의 정보를 저장하고 다른 프로세스를 실행 시킬때 발생한다. CPU는 하나의 프로세스 정보만 기억할 수 있기 때문에, 다른 프로세스를 실행하기 위해서는, 기존에 실행중이던 프로세스의 상태 혹은 레지스터 값(Context)을 PCB에 저장한 해야함. 기존 값을 저장 후, 다음 실행 시킬 프로세스의 상태와 레지스터값을 PCB에서 읽어와 레지스터에 적재하는 작업을 Context Switching이라고 한다.

5. kernel mode에서 user mode로 변경

**context switching이 발생하는 상황**

context switching은 인터럽트 발생시에 발생, 인터럽트는 실행중인 CPU사용 할당시간을 모두 소모, 입출력을 위해 대기하는 경우에 발생하며, 프로세스의 CPU제어권이 다른 프로세스에게 이양되는 상황으로 볼 수 있다. 따라서, 프로세스 상태와도 관련이 있으며, 3가지 상태 변화에 따라 인터럽트가 발생한다. CONTEXT SWITCG를 하는 주체는 OS 스케쥴러이다.

1. running > ready
2. ready > running
3. running > waiting

**PCB(Process Control Block)**

프로세스를 관리하는데 있어 필요한 정보(메타 데이터)를 담고있는 운영체제 커널의 자료구조이다. 프로세스가 생성될 때마다 PCB가 생성되며, 프로세스가 완료되면 제거된다. PCB에는 [Pocess ID, Process State(ex. ready, running), Program Counter(다은 instruction), Resister 값, Pointer, CPU scheduling/Memory/Process information]등의 정보가 포함
