# **TCP와 UDP의 헤더차이**

**UDP segment의 간략한 구조**

애플리케이션 데이터는 UDP 데이터그램의 데이터 필드에 위치. UDP헤더는 2바이트(16비트)씩 구성된 4개의 필드를 가짐. UDP헤더는 출발지 포트번호, 목적지 포트번호, 체크섬, 길이로 이뤄짐

image.png

* **포트번호**는 (목적지) 호스트가(역 다중화 기능을 수행하는) 정확한 프로세스에게 애플리케이션 데이터를 넘기게 하기 위해 사용
* **체크섬**(checksum)은 세그먼트에 오류가 발생했는지 검사하기 위해 사용되며, 체크섬은 UDP 세그먼트 이외에 IP 헤더의 일부 필드도 계싼한다.(가상 헤더)UDP 헤더와 데이터를 모두 포함하여 체크하게 됨.
* **길이**는 헤더를 포함하는 UDP 세그먼트의 길이(바이트 단위)를 나타냄. UDP헤더와 데이터를 합친 길이

**TCP segment의 간략한 구조**

TCP 소켓은 4개의 다른 요소들의 집합에 의해 식별(출발지 IP, 출발지 포트번호, 목적지 IP, 목적지 포트번호)

IP를 제외한 출발/도착지 포트번호(16bit)와 sequence number(32bit), ack number(32bit)를 합쳐, 기본적으로 20바이트의 헤더를 가지며, 옵션을 포함하면 최대 60byte의 헤더를 가질 수 있다. 다른 출발지 주소를 가지게 되는 세그먼트는 다른 소켓을 통해 프로세스에 전달됨. UDP와 다르게, TCP 세그먼트는 출발지 주소가 다르면, 다른소켓으로 전달됨

image.png

* **포트번호**는 IP정보와 결합해 출발지, 도착지를 구분하기 위해 사용
* **Sequence Number**는 SYN패킷을 보낼때, 동기화를 위해 사용되는 번호이다. 초기 Seqeunce Number를 ISN이라 부르며, 여기에는 랜덤한 함수가 담김
* **Ack Number**는 ACK패킷을 보낼 때 동기화를 위해 사용되는 번호
