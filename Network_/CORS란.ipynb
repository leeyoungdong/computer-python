# **CORS란?**

교차 출처 리소스 공유(Cross-Origin Resource Sharing)는 추가 HTTP헤더를 사용하여, 한 출처에서 실행 중인 웹 애플리케이션이 다른 출처가 선택한 자원에 접근할 수 잇는 권한을 부여하도록 브라우저에게 알려주는 체제

CORS체제는 브라우저와 서버간의 안전한 교차 출처 요청 및 데이터 전송을 지원. 최신 브라우저는 XMLHTTPRequest 또는 Fetch와 같은 API에서 CORS를 사용해 교차출처 HTTP요청을 의 위험을 완화

**CORS의 동작원리**

1. 웹은 다른 출처의 리소스를 요청할때, HTTP 프로토콜을 사용하여 요청하는데, 이때 브라우저는 요청 헤더(Request Header)에 Origin필드에 욫어을 보내는 출처를 담아 전송
2. 서버는 요청에 대한 응답을 하는데, 응답 헤더(response header)에 Access-Contrl-Allow-Origin이라는 값에 '이 리소스를 접근하는 것이 허용된 출처'를 내려 줌
3. 응답을 받은 브라우저는 자신이 보냈던 요청의 Origin과 서버가 보내준 Access-Contrl-Allow-Origin을 비교해본 후 이 응답이 유효한 응답인지를 결정
